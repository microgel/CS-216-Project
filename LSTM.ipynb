```python
# importing required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
from matplotlib import pyplot
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
```

```python
# ignore notebook warnings
warnings.simplefilter('ignore')
```

```python
# predicting AAPL stock price
df = pd.read_csv("data/AAPL.csv").set_index('Date').drop(columns=['High','Low','Open','Volume','Adj Close'])
```

```python
# number of data points
df.shape[0]
```

```python
# create dataset
dataset = df.values
```

```python
# create train and test dataset
train = dataset[0:6559,:]
valid = dataset[6559:,:]
```

```python
# use MinMaxScaler to scale data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)
```

```python
# convert dataset into x_train and y_train
x_train, y_train = [], []

# using a 60 day look back period
for i in range(60,len(train)):
    x_train.append(scaled_data[i-60:i,0])
    y_train.append(scaled_data[i,0])
    
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
```

```python
# create LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(units = 50, return_sequences = True))
model.add(LSTM(units = 50, return_sequences = True))
model.add(LSTM(units = 50, return_sequences = False))

model.add(Dense(1))
```

```python
# fit LSTM network
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train, y_train, epochs=5, batch_size=32)
```

```python
# predicting stock price with 60 day look back period from the train data
inputs = df[len(df) - len(valid) - 60:].values

inputs = inputs.reshape(-1,1)
inputs  = scaler.transform(inputs)
inputs
```

```python
# test dataset
X_test = []
for i in range(60,inputs.shape[0]):
    X_test.append(inputs[i-60:i,0])
    
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
```

```python
# closing price
closing_price = model.predict(X_test)
closing_price = scaler.inverse_transform(closing_price)
```

```python
# root mean square value
rms = np.sqrt(np.mean(np.power((valid-closing_price),2)))
rms
```

```python
# set up dataframe for visualization
train = df[:6559]
valid = df[6559:]
valid['Predictions'] = closing_price
```

```python
# plot train data, test data, predicted values
fig, ax = plt.subplots(figsize=(30,15))
ax.plot(train['Close'], label="Train")
#ax.plot(valid['Close'], label="Actual")
ax.plot(valid['Predictions'], label="Predictions")
leg = ax.legend()
```
